```python
import pandas as pd
```

```python
df_no_inf = pd.read_csv("results/concept-size-distribution/concept-distribution.tsv", sep='\t',low_memory=True, engine='python')
df_with_inf = pd.read_csv("results/concept-size-distribution/concept-distribution-inference.csv",sep=';~;',low_memory=True,engine='python')
df_mapped_concepts = pd.read_csv("results/concept-size-distribution/mapped-concepts-distribution.csv", sep=";")
```

```python
df_no_inf.head(3)
```

```python
df_with_inf.head(3)
```

```python
df_mapped_concepts.head(3)
```

```python
df_no_inf_sorted =  df_no_inf.sort_values(by='nb_instances', ascending=False)
df_no_inf_sorted.head(5)
```

```python
df_with_inf_sorted =  df_with_inf.sort_values(by='nb_instances', ascending=False)
df_with_inf_sorted.head(5)
```

```python
bins = pd.cut(df_no_inf_sorted['nb_instances'], [0, 1, 100, 1000, 100000, 100000000, 2000000000])
df_no_inf_sorted.groupby(bins)['nb_instances'].agg(['count', 'sum'])
```

```python
bins = pd.cut(df_with_inf_sorted['nb_instances'], [0, 1, 100, 1000, 100000, 100000000, 2000000000])
df_with_inf_sorted.groupby(bins)['nb_instances'].agg(['count', 'sum'])
```

```python
df_mapped_concepts_with_inf = df_mapped_concepts[df_mapped_concepts['inference']=='(with']
df_mapped_concepts_no_inf = df_mapped_concepts[df_mapped_concepts['inference']=='(no']

df_mapped_concepts_with_inf_sorted =  df_mapped_concepts_with_inf.sort_values(by='nb_instances', ascending=False)
df_mapped_concepts_no_inf_sorted =  df_mapped_concepts_no_inf.sort_values(by='nb_instances', ascending=False)

df_mapped_concepts_no_inf_sorted.head(5)
```

```python
bins = pd.cut(df_mapped_concepts_no_inf_sorted['nb_instances'], [0, 1, 100, 1000, 100000, 100000000, 2000000000])
df_mapped_concepts_no_inf_sorted.groupby(bins)['nb_instances'].agg(['count'])
```

```python
bins = pd.cut(df_mapped_concepts_with_inf_sorted['nb_instances'], [0, 1, 100, 1000, 100000, 100000000, 2000000000])
df_mapped_concepts_with_inf_sorted.groupby(bins)['nb_instances'].agg(['count'])
```
